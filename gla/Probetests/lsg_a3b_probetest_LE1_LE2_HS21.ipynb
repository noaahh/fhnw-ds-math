```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xx = np.array([[-2],[-1],[0],[1],[2],[3]])
print(xx)
```

```python
yy = np.array([[-1],[-1],[0],[0],[1],[1]])
print(yy)
```

```python
xx**4
```

```python
A=np.hstack((xx**5,xx**4,xx**3,xx**2,xx,xx**0))
print(A)
```

```python
aa = np.dot(np.linalg.inv(A),yy)
print(aa)
```

```python
plt.plot(xx,yy,'r*')
```

```python
x = np.arange(-2,3,0.01)
print(x)
y = aa[0]*x**5+aa[1]*x**4+aa[2]*x**3+aa[3]*x**2+aa[4]*x+aa[5]*x**0
print(y)
```

```python
plt.plot(x,y,'-b',xx,yy,'r*')
```
